```python
import numpy as np
from matplotlib import pyplot as plt

def vector_length(coords):
    if len(coords) != 3:
        raise "This is not a vector"
    return np.sqrt(coords[0]**2 + coords[1]**2 + coords[2]**2)

vector_length([3, 2, 1])
```

### 4. Задание (на листочке)
**1) Пусть задана плоскость:**

$$Ax + By + Cz + D = 0$$

Напишите уравнение плоскости, параллельной данной и проходящей через начало координат.


#### Решение:

Из уравнения плоскости следует, что коэффициент D плоскости, проходящей через начало координат должен быть равен нулю.
Так же из определения уравнения плоскости, коэффициенты A, B , и C характеризуют тангенсы углов наклона к соответствующим осям. И для сохранения углов достаточно сохранить коэффициенты.

#### Ответ: 
$$Ax + By + Cz = 0$$

#### Проверка.
$$A*0 + B*0 + C*0 = 0$$
Плоскость проходит через начало координат.

\begin{cases}Ax + By + Cz = D
\\Ax + By + Cz = 0\end{cases}

Очевидно, что система из двух уравнений плоскостей не имеет решений при D != 0, что значит плоскости не пересекаются, и, по определению, параллельны.

### 3. Задание

#### Напишите код, который будет переводить полярные координаты в декартовы.

Полярная система координат:
$$x = R\cdot cos\alpha$$
$$y = R\cdot sin\alpha$$

```python
def polar_to_decard(r, alpha):
    if r < 0:
        raise Exception("R should be >= 0")
    if alpha <= 0 or alpha >= 2*np.pi:
        raise Exception("alpha should be between 0 and 2pi")
    x = r * np.cos(alpha)
    y = r * np.sin(alpha)
    return x, y

polar_to_decard(2, 1.3*np.pi)
```

### 4. Задание (в программе)

Решите систему уравнений:
$$
\begin{cases}
y = x^2 - 1 \\
exp(x) + x(1 - y) = 1
\end{cases}
$$


```python
from scipy.optimize import fsolve
```

$$
\begin{cases}
x^2 - 1 - y = 0\\
exp(x) + x - xy - 1 = 0
\end{cases}
$$


```python
def equations(p):
    x, y = p
    return  x**2 - 1 - y, np.exp(x) + x - x*y - 1
```

```python
x = np.linspace(-2,4.7, 300)
plt.plot(x, x**2 - 1)
plt.plot(x, (np.exp(x) + x - 1) / x)
plt.grid(True)
```

```python
x1, y1 = fsolve(equations, (-2, 1))
(x1, y1)
```

```python
x2, y2 = fsolve(equations, (2, 5))
(x2, y2)
```

```python
x3, y3 = fsolve(equations, (7, 30))
(x2, y2)
```
